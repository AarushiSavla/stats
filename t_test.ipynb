```python
import numpy as np
import scipy.stats as stats 
```

```python
sample=[172,174,168,169,171,173,175,170,169,172]
df=9
sample_mean = np.mean(sample)
population_mean= 170
sample_Std= np.std(sample,ddof=1)
n=len(sample)
```

```python
t_test= (sample_mean-population_mean)/(sample_Std/np.sqrt(n))
t_test
```

```python
alpha = 0.05
```

```python
p_val = 2 * (1 - stats.t.cdf(abs(t_test), df))
p_val
```

```python
if p_val<alpha:
    print("reject the null hypothesis")
else :
    print("accept the null hypothesis")
```
