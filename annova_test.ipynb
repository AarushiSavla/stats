```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
from scipy.stats import f_oneway
```

```python
df= sns.load_dataset('titanic')
df
```

```python
df=df[['age','pclass']].dropna()
df
```

```python
df['pclass'].unique()
```

```python
class1 =df[df['pclass']==1]['age']
class2 =df[df['pclass']==2]['age']
class3 =df[df['pclass']==3]['age']
class1
```

```python
h0 = " there is no differnce between the passenger class and ages"
h1 ="there is a differnce  between the passenger class and ages "
h0,h1
```

```python
f_stats, p_value = f_oneway(class1,class2,class3)
```

```python
f_stats
```

```python
p_value
```

```python
alpha = 0.05 
if p_value<alpha :
    print(h1)
else:
    print(h0)
```
